```python
import numpy as np
import matplotlib.pyplot as plb
```

```python
t = np.linspace(0,40,4000) #start,finish,n points
```

```python
drobber = 2.5 * t
dcop = 3 * (t-5)
```

```python
fig, ax= plb.subplots()
plb.title("when will the robber get caught?")
plb.ylabel("disance")
plb.xlabel('time (in mins)')
ax.set_xlim([0,40])
ax.set_ylim([0,100])
ax.plot(t,drobber, c='green')
ax.plot(t,dcop, c ='red')
plb.axvline(x=30, color='purple', linestyle='--')
plb.axhline(y=75, color='purple', linestyle='--')
```
